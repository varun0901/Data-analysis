```python
#songs for different genres

spotify_feature = pd.read_csv("C:/Users/asus/OneDrive/Desktop/data_analyst/Spotify_features/SpotifyFeatures.csv")

data = pd.DataFrame(spotify_feature)

artist_name = data["artist_name"].head(20)
danceability = data["danceability"].head(20)

sb.barplot (x = danceability, y = artist_name).set_title (" Artist name and dancebility of their songs")
```
